```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
  return np.sum(x**2)
#parameter
n_agents = 30
dim = 30
max_iter = 100
lb = -10
ub = 10
#coffecients
c_max = 1
c_min = 0.0004

def S_function(r):
  f= 0.5
  l = 1.5
  return f * np.exp(-r / l) - np.exp(-r)

position = np.random.uniform(lb, ub, (n_agents, dim))
best_pos = np.zeros(dim)
best_score = float("inf")

converengence_curve = []
for t in range(max_iter):
  c = c_max -t * (c_max - c_min) / max_iter

  for i in range(n_agents):
    temp_pos = np.zeros(dim)
    for j in range(n_agents):
      if  i != j:
        r = np.linalg.norm(position[i] - position[j])
        unit_vec = (position[i] - position[j]) / (r + 1e-8)
        temp_pos += c *((ub - lb)/2) * S_function(r) * unit_vec
    new_pos = c* temp_pos + best_pos
    new_pos = np.clip(new_pos, lb, ub)
    position[i] = new_pos
    fitness = objective_function(new_pos)
    if fitness < best_score:
      best_pos = new_pos
      best_score = fitness
  converengence_curve.append(best_score)
  if t % 10 ==0 or t == max_iter -1:
    print(f"Iteration {t+1}: Best Score = {best_score}")

plt.plot(converengence_curve)
plt.xlabel("Iteration")
plt.ylabel("Best Score")
plt.title("Convergence Curve")
plt.grid()
plt.show()
print(f"Best Position: {best_pos}")
print(f"Best Score: {best_score}")
```
